```python
#this will discard all samples other than the chosen high and low producers
results2=pd.merge(left=results,right=ctype,left_index=True,right_index=True,how="left")
```

```python
X = results2.iloc[:,:-1].values
y = results2["type"]

#X=RobustScaler().fit_transform(X)
#X=MinMaxScaler(feature_range=(0,1)).fit_transform(X)
#X=StandardScaler().fit_transform(X)
```

```python
fit = umap.UMAP(n_neighbors=100,min_dist=1)
%time u = fit.fit_transform(X)
```

```python
principalDf = pd.DataFrame(data = u
             , columns = ['UMAP 1-D', 'UMAP 2-D'],index=results2.index)
```

```python
finalDf = pd.concat([principalDf, results2['type']],axis=1)
```

```python
finalDf.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('UMAP 1-D', fontsize = 15)
ax.set_ylabel('UMAP 2-D', fontsize = 15)
ax.set_title('2D UMAP', fontsize = 20)
targets = ['High_producer', 'Low_producer']
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['type'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'UMAP 1-D']
               , finalDf.loc[indicesToKeep, 'UMAP 2-D']
               , c = color
               , s = 50,alpha=0.4)
ax.legend(targets)
ax.grid()
plt.tight_layout()
ax.figure.savefig("UMAP_40_qnorm_nos.png",bbo_inches="tight")
```
